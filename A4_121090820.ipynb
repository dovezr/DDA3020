```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_table(r"C:\Users\86180\Downloads\seeds_dataset.txt",sep='\s+',header = None)
x = data[data.columns[0:7]]
y = data[7]
```

```python
#PCA
x = np.mat(x)
ecm = 1/210*np.dot((x-np.mean(x,axis = 0)).transpose(),(x-np.mean(x,axis = 0)))#claculate the empirical covariance matrix

#print(ecm)
eigenvalue, eigenvector = np.linalg.eig(ecm)


K_index = np.argsort(eigenvalue)



K_index = K_index[:-3:-1]

eigenvector = eigenvector[:,K_index]

new_x = np.dot((x-np.mean(x,axis = 0)),eigenvector)

print(new_x)
```

```python
def distance(a,b):
    return np.sqrt((a[0,0]-b[0,0])**2+(a[0,1]-b[0,1])**2)
```

```python
#K-means
centroid = [new_x[0,:],new_x[1,:],new_x[2,:]]
label = [0]*210
while(True):
    new_label = [0]*210
    x1=0;x2=0;x3=0;y1=0;y2=0;y3=0
    num1=0;num2=0;num3=0
    list0=[];list1=[];list2=[]
    for i in range(210):
        d1 = distance(new_x[i,:],centroid[0])
        d2 = distance(new_x[i,:],centroid[1])
        d3 = distance(new_x[i,:],centroid[2])
        if d1 < d2 and d1 < d3:
            new_label[i] = 0
            x1+=new_x[i,0]
            y1+=new_x[i,1]
            num1+=1
            list0.append(i)
        elif d2 < d1 and d2 < d3:
            new_label[i] = 1
            x2+=new_x[i,0]
            y2+=new_x[i,1]
            num2+=1
            list1.append(i)
        else:
            new_label[i] = 2
            x3+=new_x[i,0]
            y3+=new_x[i,1]
            num3+=1
            list2.append(i)
    if new_label == label:
        break
    label = new_label
    centroid = [np.array([[x1/num1,y1/num1]]),np.array([[x2/num2,y2/num2]]),np.array([[x3/num3,y3/num3]])]
print(label)
```

```python
#silhouette coefficient
def sc(a,b):
    return ((b-a)/max(a,b))
```

```python
s = []
for i in range(210):
    if label[i] == 0:
        sum0 = 0
        for j in list0:
            if i != j:
                sum0+=distance(new_x[i,:],new_x[j,:])
        sum1 = 0
        sum2 = 0
        for j in list1:
            sum1+=distance(new_x[i,:],new_x[j,:])
        for j in list2:
            sum2+=distance(new_x[i,:],new_x[j,:])    
        a = sum0/(len(list0)-1)
        b = min(sum1/len(list1),sum2/len(list2))
        s.append(sc(a,b))
    elif label[i] == 1:
        sum0 = 0
        for j in list1:
            if i != j:
                sum0+=distance(new_x[i,:],new_x[j,:])
        sum1 = 0
        sum2 = 0
        for j in list0:
            sum1+=distance(new_x[i,:],new_x[j,:])
        for j in list2:
            sum2+=distance(new_x[i,:],new_x[j,:])    
        a = sum0/(len(list1)-1)
        b = min(sum1/len(list0),sum2/len(list2))
        s.append(sc(a,b))
    else:
        sum0 = 0
        
        for j in list2:
            if i != j:
                sum0+=distance(new_x[i,:],new_x[j,:])
        sum1 = 0
        sum2 = 0
        for j in list1:
            sum1+=distance(new_x[i,:],new_x[j,:])
        for j in list0:
            sum2+=distance(new_x[i,:],new_x[j,:])    
        a = sum0/(len(list2)-1)
        b = min(sum1/len(list0),sum2/len(list1))
        s.append(sc(a,b))
print("The silhouette coefficient is",sum(s)/len(s))
```

```python
# Rand index
tol = len(list0+list1+list2)
same=0;dif=0
for i in range(210):
    for j in range(i+1,210):
        if label[i]==label[j] and y[i]==y[j]:
            same+=1
        if label[i]!=label[j] and y[i]!=y[j]:
            dif+=1
RI = (same+dif)/((tol*(tol-1))/2)
print("The rand index is", RI)
```
